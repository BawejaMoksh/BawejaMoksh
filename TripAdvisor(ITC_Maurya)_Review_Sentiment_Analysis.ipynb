<h1><b>Sentiment Analysis of Trip Advisor review through TextBlob model

```python
import requests as rq
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
```

```python
r = rq.get('https://www.tripadvisor.in/Hotel_Review-g304551-d301777-Reviews-ITC_Maurya-New_Delhi_National_Capital_Territory_of_Delhi.html')
```

```python
r.status_code
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text,'html.parser')
```

```python
spans = soup.findAll(class_='QewHA H4 _a')
```

```python
reviews = []
for span in spans:
  reviews.append(span.find('span').text)
```

```python
print(reviews,'\n')
```

```python
df = pd.DataFrame(np.array(reviews),columns =['review'])
```

```python
df.head()
```

```python
len(df['review'])
```

```python
df['review'].apply(lambda x: len(x.split()))
```

```python
df['review'].apply(lambda x: len(x))
```

<h1><b>Data Analysis

```python
df['word_count']= df['review'].apply(lambda x: len(x.split()))
```

```python
df['char_count'] = df['review'].apply(lambda x: len(x))
```

```python
df.head()
```

```python
def avg_word(x):
  word = x.split()
  return sum(len(w) for w in word)/len(word)
```

```python
df['avg_charl'] = df['review'].apply(lambda x:avg_word(x))
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
st_w = stopwords.words('english')
```

```python
st_w
```

```python
df['review'].apply(lambda x: len(([w for w in x.split() if w.lower() in st_w])))
```

```python
df['stop_word_count'] = df['review'].apply(lambda x: len(([w for w in x.split() if w.lower() in st_w])))
```

```python
df['stop_word_rate']=df['stop_word_count']/df['word_count']
```

```python
df.head()
```

```python
df.sort_values(by='stop_word_rate')
```

```python
df.describe()
```

<h1><b>Data Cleaning

```python
df['lower_review']=df['review'].apply(lambda x: x.lower())
```

```python
df['punctuation']=df['lower_review'].str.replace('[^\w\s]','')
```

```python
df['stopword']=df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in st_w))
```

```python
pd.Series(' '.join(df['stopword']).split()).value_counts()[:35]
```

```python
other_st_w = ['also','us','entire','get','always','reallu','enough']
```

```python
df['clean'] = df['stopword'].apply(lambda x: " ".join(w for w in x.split() if w not in other_st_w))
```

```python
df.head()
```

<h1><b>Lemmatizing of Data

```python
import textblob
from textblob import Word
nltk.download('wordnet')
```

```python
df['clean'].apply(lambda x: " ".join(Word(w).lemmatize() for w in x.split())) 
```

```python
df.head()
```

<h1><b>Sentiment Analysis

```python
from textblob import TextBlob
```

```python
df['polarity']=df['clean'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity']=df['clean'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(['lower_review','punctuation','stopword','clean'],axis=1,inplace = True)
```

```python
df.head()
```

```python
df.sort_values(by='polarity')
```
